```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
import pickle
from sklearn.externals import joblib
```

```python
df=pd.read_csv("insurance.csv")
df.head()
```

```python
df.describe()
```

```python
 value=np.random.rand(len(df))<.8
```

```python
train=df[value]
test=df[~value]
```

```python
regr=lm.LinearRegression()
train_x=np.asanyarray(train[["age","bmi","children"]])
train_y=np.asanyarray(train[["charges"]])
regr.fit(train_x,train_y)
```

```python
print(regr.coef_)
```

```python
print(regr.intercept_)
```

```python
joblib.dump(regr,"insurance.pkl")
```

```python
test_x=np.asanyarray(test[["age","bmi","children"]])
test_y=np.asanyarray(test[["charges"]])
prediction=regr.predict(test_x)
print(prediction)
```
